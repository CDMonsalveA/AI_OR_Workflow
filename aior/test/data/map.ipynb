```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
locations = pd.read_csv('location_info.csv')
# make a map of the locationswith the 'lat' and 'lon' columns, and hover text of 'name'
fig = px.scatter_mapbox(locations, lat="lat", lon="lon", hover_name="name", hover_data=["name", "lat", "lon"],
                        zoom=4)
# Make a line connecting the cities, and customize the line color and thickness
# update the layout to expand the map to full screen
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
# show the map
fig.show()
# save the map as an html file
fig.write_html("map.html")
```
