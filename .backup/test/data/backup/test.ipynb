```python
import pandas as pd

poblacion = pd.read_csv('poblacion_por_municipio_total.csv')
#  columns = dp	dpnom	dpmp	mpio	1985	1986	1987	1988	1989	1990	...	2026	2027	2028	2029	2030	2031	2032	2033	2034	2035
comida_por_persona = 1.3/1000

poblacion['demanda'] = poblacion['2024']*comida_por_persona

# Constantes
costo_por_m2 = 24450 / 4 # COP/semana

densidad_de_area_de_alimentos = 10.5/7.8 * 1.1 # ton/m2
# Demanda (Ton/semana) <- Ya lo tengo

poblacion['costo_de_facility'] = poblacion['demanda'] * costo_por_m2 / densidad_de_area_de_alimentos # COP/semana

demanda_por_departamentos = poblacion.groupby(['dp', 'dpnom']).sum()['demanda']
display(demanda_por_departamentos)

display(poblacion[['mpio', '2024', 'demanda', 'costo_de_facility']])


# Costo de transporte es  3364/10.5 COP/(ton*km)
```
